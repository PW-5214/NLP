```python
sentences = df['reviewText'].dropna().tolist()
```

```python
# Preprocessing
import nltk
import re
from nltk.tokenize import word_tokenize
from nltk.corpus import stopwords

nltk.download('punkt')
nltk.download('stopwords')

stop_words = set(stopwords.words('english'))

def preprocess(text):
    text = text.lower()
    text = re.sub(r'[^a-z\s]', '', text)
    words = word_tokenize(text)
    return [w for w in words if w not in stop_words]

tokenized_sentences = [preprocess(s) for s in sentences]
print(tokenized_sentences[0])
```

```python
# model Word2vec
from gensim.models import Word2Vec

model = Word2Vec(
    sentences=tokenized_sentences,
    vector_size=100,
    window=5,
    min_count=5
)
```

```python
output = model.wv.most_similar("movie")

for word, score in output:
    print(word, score)
```

```python
model.wv.similarity("movie","movies")
```

```python
# model Fast - texting
from gensim.models import FastText

fasttext_model = FastText(
    sentences=tokenized_sentences,
    vector_size=100,
    window=5,
    min_count=5
)
```

```python
output = fasttext_model.wv.most_similar("movie","movies")

for word, score in output:
    print(word, score)
```
